# Image Mirror with Python.

```python
from PIL import Image
```

```python
Image.open('Pythontest.png') # Original Image
```

```python
img = Image.open('Pythontest.png')
Mirror_Image=img.transpose(Image.FLIP_LEFT_RIGHT)
Mirror_Image.save(r'Pythontest_mirror.png')
Image.open('Pythontest_mirror.png')
```
